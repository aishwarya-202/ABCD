```python
le_col=['Client Code', 'Location Code', 'Location Name', 'Policy No.',
       'Claim No.', 'Claimant', 'Close Date', 
       'Accident Desc',  'Risk State', 'Acc State',
       'Jurisdiction', 'Adjuster',
       'Cause Code', 'Cause Desc', 'Injury Type', 'Accident City',
       'Accident State']
```

```python
from sklearn.preprocessing import LabelEncoder
le= LabelEncoder()
```

```python
for col in le_col:
   df_encoded[col]=le.fit_transform(df_encoded[col])
```

```python
df_encoded.info()
```

```python

X= df_encoded.drop(columns=["Total Inc","Inception","Expiration","Occ Date","Report Date","Receive Date"])
X=sm.add_constant(X)
y= df_encoded["Total Inc"]
model=sm.OLS(y,X).fit()
print(model.summary())
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

vif= pd.DataFrame()
vif["att"]=X.columns
vif["vif"]=[variance_inflation_factor(X.values,i) for i in range(X.shape[1])]
vif.sort_values(by="vif", ascending = False, inplace = True, ignore_index= True)
while vif["vif"][0]>5:
    X.drop(vif["att"][0],axis=1,inplace=True)
    vif=pd.DataFrame()
    vif['att']=X.columns
    vif["vif"]=[variance_inflation_factor(X.values,i) for i in range(X.shape[1])]
    vif.sort_values(by="vif", ascending = False, inplace = True, ignore_index= True)
```

```python
latest_date=df_encoded["Report Date"].max()
three_years_ago= latest_date- pd.DateOffset(years=3)
filtered_df= df_encoded[df_encoded["Report Date"] >= three_years_ago]
```

```python
filtered_df
```

```python
X=filtered_df[['Location Name', 'Close Date', 'Adjuster', 'Accident Desc',
       'Accident City', 'Claimant', 'Risk State', 'Client Code',
       'Exp Paid', 'Exp Resv', 'Ind/BI Paid', 'Litigated_Y',
       'Claim Type_Indemnity', 'Ind/BI Resv', 'APD Reco',
       'Claim Type_Medical Only', 'Status_O', 'Med/PD Paid',
       'APD Exp Paid', 'Claim Type_Bodily Injury', 'Subrogated_Y',
       'APD Resv', 'APD Inc', 'Med/PD Resv', 'APD Exp Resv',
       'Carrier Name_ACE Fire Underwriters Insurance Company', 'Recovery',
       'Status_R', 'Claim Type_XMO', 'Claim Type_Record Only']]
X=sm.add_constant(X)
y= filtered_df["Total Inc"]
model=sm.OLS(y,X).fit()
print(model.summary())
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2, random_state=10)
model=sm.OLS(y_train,X_train).fit()
model.summary()
```

```python
y_pred_test=model.predict(X_test)

from sklearn.metrics import mean_squared_error, r2_score
print("MSE", mean_squared_error(y_test,y_pred_test))
print("R_square Score:", r2_score(y_test,y_pred_test))
```

```python
y_pred_train=model.predict(X_train)

from sklearn.metrics import mean_squared_error, r2_score
print("MSE", mean_squared_error(y_train,y_pred_train))
print("R_square Score:", r2_score(y_train,y_pred_train))
```

```python
from statsmodels.graphics.gofplots import qqplot

qqplot(model.resid,line="r")
plt.show()
```
